```python
data_sel = data[['ID', 'POI_NM', 'LC_LA', 'LC_LO']]
data_sel.head()
```

```python
# db 삽입 전 테스트
destination = (input('도착지를 입력해주세요(출발지는 임의로 설정됨): ')).strip() 
# 이름만으로 처리할 때 중복된 값이 있음 e.g.) 안정사 7개 # test: 약사원

lat = 0; lon = 0; place = ''
if destination in data_sel['POI_NM'].values:
    idx = np.where(data_sel['POI_NM']==destination)
    lat = data_sel['LC_LA'].iloc[idx]
    lon = data_sel['LC_LO'].iloc[idx]
    place = data_sel['POI_NM'].iloc[idx]
else: print('NOT EXIST')

from haversine import haversine
start = (35.72930553, 128.9320902) # (latitude, longitude) # start point 임의로 설정
end = (lat.all(), lon.all())
haversine(start, end, unit='km') # unit에 다른 단위로 설정 가능
```

```python
# db 삽입
import sqlite3

conn = sqlite3.connect('./database.db')
cursor = conn.cursor()
try:
    sql = "CREATE TABLE IF NOT EXISTS location_tb (id TEXT PRIMARY KEY, place TEXT, lat REAL, lon REAL);"
    cursor.execute(sql)
    for val in data_sel.values:
        sql = "INSERT INTO location_tb VALUES(?, ?, ?, ?);"
        cursor.execute(sql, [val[0], val[1], val[2], val[3]])
        conn.commit()
except:
    pass
finally:
    conn.close()
```

```python
# db에서 테스트
import sqlite3

conn = sqlite3.connect('./database.db')
cursor = conn.cursor()

start = input('출발장소를 선택하세요: ')
sql = "SELECT * FROM location_tb WHERE place = '%s';" %start
cursor.execute(sql)
start_point = cursor.fetchall()
(start_id, start_place, start_lat, start_lon) = start_point[0]

destination = input('도착지 장소를 선택하세요: ')
sql = "SELECT * FROM location_tb WHERE place = '%s';" %destination
cursor.execute(sql)
destination_point = cursor.fetchall()
(dest_id, dest_place, dest_lat, dest_lon) = destination_point[0]

distance = haversine((start_lat, start_lon),(dest_lat, dest_lon), unit='km')
print(distance)

conn.close()
```
